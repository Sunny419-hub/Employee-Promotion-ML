4. Elimina los registros con información faltante (`NaN`) y la columna `employee_id` por no tener aportación en la predicción.

```python
# Revisión de valores faltantes por columna
print("Valores faltantes por columna:")
print(info.isnull().sum())
```

```python
# Elimina filas con valores faltantes
clean = info.dropna(axis=0).copy()
```

5. Codifica las variables categóricas de info con `LabelEncoder`.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# dataframe límpio

encoded = clean.copy()

#identificar variables categóricas
cat_cols = encoded.select_dtypes(include=["object", "category"]).columns.tolist()
print("Variables categóricas:", cat_cols if cat_cols else "— (no hay)")
```

```python
# Codificar con LabelEncoder
label_encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    encoded[col] = le.fit_transform(encoded[col].astype(str))
    label_encoders[col] = le

# Vista después de codificar
print("Shape tras codificación:", encoded.shape)
encoded.head()
```

6. Divide la matriz en conjuntos de entrenamiento y prueba (80:20), utilizando el parámetro `random_state=1` para garantizar reproducibilidad.

```python
from sklearn.model_selection import train_test_split
from collections import Counter
```

```python
# Divide el conjunto dataframe info para obtener: train y test
X = encoded.drop(columns=["is_promoted"])
y = encoded["is_promoted"]
```

```python
#Conjuntos de entrenamiento y prueba (80:20) con random_state=1

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=1,
    stratify=y
)

print("Distribución original:", Counter(y))
print("Train:", Counter(y_train), " Test:", Counter(y_test))
print("Shapes -> X_train:", X_train.shape, " X_test:", X_test.shape)
```

7. Observa la distribución de la variable objetivo `is_promoted`. En caso de que las clases estén desbalanceadas, ejecuta `over_sampling` con el método `SMOTE()`.

```python
# Observa la distribución (cuántos empleados fueron promovidos (1) y no promovidos (0) en el conjunto de entrenamiento)

print("Antes de SMOTE:", Counter(y_train))

```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
```

```python
# Separa los predictores (X) y la variable de salida (y)
# Calcular el porcentaje de promovidos
ratio = y_train.mean()
print(f"Porcentaje de promovidos en train: {ratio:.2%}")
```

```python
# Aplica SMOTE al conjunto de entrenamiento (train[X], train[y])
if ratio < 0.4 or ratio > 0.6:
    sm = SMOTE(random_state=1)
    X_train_bal, y_train_bal = sm.fit_resample(X_train, y_train)
    print("Después de SMOTE:", Counter(y_train_bal))
else:
    X_train_bal, y_train_bal = X_train, y_train
    print("Las clases están balanceadas, no aplicamos SMOTE.")
```

8. Configura el autoaprendizaje de PyCaret asegurándote de usar `preprocess=False` para no aplicar ninguna operación adicional al conjunto.

```python
from pycaret.classification import *
```

```python
# Configura y ejecuta el autoaprendizaje (variable objetivo SIN PREPROCESS)

clf = setup(
    data = pd.concat([X_train_bal, y_train_bal], axis=1),
    target = "is_promoted",
    session_id = 1,
    preprocess = False,
    train_size = 0.8,
    verbose = False
)
```

9. Optimiza el mejor modelo `tune_model()` con la métrica `accuracy`.

```python
from pycaret.classification import tune_model, pull
```

```python
# Comparar modelos y elegir el mejor de acuerdo a (accuracy)
best_model = compare_models(sort='Accuracy')
print("Mejor modelo seleccionado:", best_model)

# Optimizar del mejor modelo (usando accuracy)
tuned_best = tune_model(best_model, optimize='Accuracy')
print("Modelo optimizado listo.")
```

10. Utiliza `evaluate_model()` para mostrar la interfaz de usuario con detalles del modelo optimizado y verifica las métricas `predict_model() `con el conjunto de prueba.

```python
from pycaret.classification import evaluate_model, predict_model, pull
import pandas as pd
```

```python
# Detalles del modelo
evaluate_model(tuned_best)
```

```python
# Verifica las métricas (con el conjunto de PRUEBA)

# DataFrame de test con la columna objetivo 'is_promoted'
test_df = pd.concat([X_test.copy(), y_test.copy()], axis=1)
if test_df.columns[-1] != "is_promoted":
    test_df.columns = list(X_test.columns) + ["is_promoted"]

# Predicciones en TEST
test_predictions = predict_model(tuned_best, data=test_df)

# Mostrar las métricas calculadas
metrics_test = pull()
print("Métricas en el conjunto de PRUEBA:")
display(metrics_test)

# Mostrar
test_predictions.head()
```
